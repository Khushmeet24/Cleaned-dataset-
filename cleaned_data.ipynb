```python
import pandas as pd

df = pd.read_csv("India-Tourism-Statistics-2022-Table-5.1.2_0.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df["states/uts"] = df["states/uts"].str.replace("*", "", regex=False).str.strip()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
numeric_cols = df.columns[2:]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.to_csv("cleaned_tourism_data.csv", index=False)
```
